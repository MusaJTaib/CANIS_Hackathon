```python


import spacy
import pandas as pd
import csv

# Load the English language model in spaCy
nlp = spacy.load('en_core_web_md')

# Load the dataframe
df = pd.read_csv('Preprocessed_Dataset_Misinfo_FAKE.csv')

# Initialize an empty list to store entities
entities = []

# Process each row in the dataframe
for index, row in df.iterrows():
    text = str(row['text'])  # Convert the input to a string
    doc = nlp(text)

    # Iterate over each entity in the document and append it to the entities list
    for ent in doc.ents:
        entities.append((ent.label_, ent.text))


```

```python
entities
```

```python
counts = {}
for tup in entities:
    if tup in counts:
        counts[tup] += 1
    else:
        counts[tup] = 1

# with open('ner_fake_counts.txt', 'w', encoding="utf-8") as f:
#     for key, value in counts.items():
#         f.write(f"{key}: {value}\n")
```

```python
counts
```

```python
from collections import defaultdict


# Convert dictionary items to a list of tuples
entity_list = [(k[0], k[1], v) for k, v in counts.items()]

# Group the entities by their type
entities_by_type = defaultdict(list)
for item in entity_list:
    entities_by_type[item[0]].append(item)

# Sort and keep the top 100 values for each entity type
top_100_entities_by_type = {}
for entity_type, entities in entities_by_type.items():
#     top_100_entities_by_type[entity_type] = sorted(entities, key=lambda x: -x[2])[:100]
    top_100_entities_by_type[entity_type] = sorted(entities, key=lambda x: -x[2])

# Save the result to a text file
with open('sorted_entities.txt', 'w', encoding="utf-8" ) as f:
    for entity_type, entities in top_100_entities_by_type.items():
#         f.write("Number of named entities {}:{}\n".format(entity_type, len(top_100_entities_by_type[entity_type])))
        for item in entities:
            f.write(f"{item[0]}, {item[1]}, {item[2]}\n")
        f.write("\n")  # Add a blank line between different entity types
        
```

```python

# Save the result to a text file
with open('Number_named_entities_Count.txt', 'w', encoding="utf-8" ) as f:
    for entity_type, entities in top_100_entities_by_type.items():
        f.write("Number of named entities {}:{}\n".format(entity_type, len(top_100_entities_by_type[entity_type])))
        
```
